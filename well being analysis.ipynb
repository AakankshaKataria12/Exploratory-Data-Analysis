# Analysis of the trend of stress levels with respect to various factors.

dataset used: https://www.kaggle.com/datasets/ydalat/lifestyle-and-wellbeing-data

In this dataset stress levels are depicted on a scale of 0-5.

With 0 being least stressed and 5 being most stressed

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('wellbeing_dataset.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df = df[df["DAILY_STRESS"] != '1/1/00']
```

```python
df.info()
```

```python
ax = sns.countplot(x='GENDER',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

Graph showing a visual representation of the number of responses from male and female users.

```python
sns.countplot(data=df, x="GENDER", hue="DAILY_STRESS")
```

Graph depicting number of responses from male and female users and there stress levels.

```python
ax = sns.countplot(data=df,x='AGE', hue='DAILY_STRESS')
sns.set(rc = {'figure.figsize':(10,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

Graph between age-groups and count of responses with seperate colors for different stress levels.

```python
ax = sns.barplot(data=df, x='DAILY_STRESS',y='WORK_LIFE_BALANCE_SCORE')
sns.set(rc = {'figure.figsize':(10,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

Graph between work-life balance score and stress levels.

```python
ax = sns.barplot(data=df, x='DAILY_STRESS',y='SLEEP_HOURS')

for bars in ax.containers:
    ax.bar_label(bars)
```

Graph between sleep hours and stress levels.

OBSERVATION: Higher the number of sleep hours, lower the stress levels.

```python
ax = sns.barplot(data=df, x='DAILY_STRESS',y='WEEKLY_MEDITATION')

for bars in ax.containers:
    ax.bar_label(bars)
```

Graph between weekly hours of meditation and corresponding daily stress levels. 

OBSERVATION: Higher hours of meditation lead to lower stress levels.

CONCLUSION: analysis of stress level was done with respect to various factors.
